```python
import requests
from bs4 import BeautifulSoup
```

```python
url='https://www.apnikheti.com/hn/pn/कृषि/फसलें/हरी-खाद/सनई'
page=requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
a=soup.find_all('article',class_='crop-info-repeater')
for x in a:
    t=x.find('h2')
    if(t is not None):
        print("LOL "+t.text)
        f=open(t.text+ ".txt","a", encoding="utf-8")
    u=x.find('div',class_='more')
    i=i+1
    if(u is not None):
        print(u.text)
        if(f.closed):
            continue
        f.write(u.text)
        f.close()
```
